```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth


temp1, temp2, volt = np.loadtxt('dataPeltier1.txt', unpack=True)
rate = 1000;

tiempo=np.arange(0,len(temp1))*(1/rate);

ds1=smooth(temp1,300);
ds2=smooth(temp2,300);

plt.figure(1)
plt.xlabel('tiempo (s)')
plt.ylabel('Temp (C)')
plt.title('Temp 1y2')
plt.plot(tiempo,ds1*100,tiempo,ds2*100,tiempo,(ds2-ds1)*100)
plt.legend(r'$x$ $y$ $z$')
plt.show()


plt.figure(2)
plt.xlabel('tiempo (s)')
plt.ylabel('Voltaje (v)')
plt.title('Voltaje')
plt.plot(tiempo,volt)
plt.show()
```
